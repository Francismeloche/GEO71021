## Resampling de Raster

```python
import rioxarray as rio
from rasterio.plot import show
from rasterio.enums import Resampling
```

```python
xdr = rio.open_rasterio('/home/francis/Documents/UQAR/Cours/GEO71021/Tp1/canopy_JBC.tif', masked = True)

res = xdr.rio.resolution() # get the resolution of the RasterArray
        # resample data to target shape
print('resolution ',res)
```

```python
upscale_factor = res[0]/0.5 #on s'en va vers une résolution de 0.5m
print('upscale_factor',upscale_factor)
new_width = int(xdr.rio.width * upscale_factor)
new_height = int(xdr.rio.height * upscale_factor)
```

```python
%%time
xdr_resamp = xdr.rio.reproject(xdr.rio.crs, shape = (new_height, new_width), resampling = Resampling.cubic)
print(xdr_resamp.rio.resolution())
```
